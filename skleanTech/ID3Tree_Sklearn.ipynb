```python
df1 = pd.read_csv("elections_clean.csv")
df = df1[['democrat', 'county', 'state', 'education', 'religion', 'ethnic_male','ethnic_female']]
#df.drop(columns=['county'], axis = 1, inplace = True)
df.dropna(inplace = True)
df
```

```python
target = 'democrat'
y = df[target]
df.drop([target], axis = 1, inplace = True)

X = pd.get_dummies(df)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
#print(one_hot_data)
```

```python
ID3Tree = DTC(criterion= 'entropy')
ID3Tree.fit(X_train, y_train)

predictions = ID3Tree.predict(X_test)
#print(y_test)
print("Accuracy (%):",metrics.accuracy_score(y_test, predictions)*100,"%")

```

```python
print(tree.export_text(ID3Tree))
```
